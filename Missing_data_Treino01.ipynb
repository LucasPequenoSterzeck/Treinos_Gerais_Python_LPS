```python
# Utilizando a própria função do Pandas, é possível orientar quais valores estão corrompidos

missing_value_formats = ["n.a.","?","NA","n/a", "na", "--"]

df = pd.read_csv("https://raw.githubusercontent.com/ChaitanyaBaweja/Programming-Tutorials/master/Missing-Data-Pandas/employees.csv",
                 na_values = missing_value_formats)

print(df['Gender'].head(10))
```

```python
gender_na = df["Gender"].isnull().sum()
salary_na = df["Salary"].isnull().sum()

print(f"Possuímos na coluna Gender {gender_na} valores em branco/corrompidos.")
print(f"Possuímos na coluna Salary {salary_na} valores em branco/corrompidos.")
```

```python
mean_Salary = df["Salary"].mean()

# Variavel que tenta transformar elemento em inteiro, porém se ele for NaN, retorna  NaN
def seja_inteiro(i):
  try:
    return int(i)
  except:
    return pd.np.nan

# Aplicando mapeamento na coluna reescrevendo-a
df["Salary"] = df["Salary"].map(seja_inteiro)
print(df.head(3),"\n\n",df.dtypes)
```

```python
#Substituindo Salary NaN por mediana

Salary_median = df["Salary"].median()

df["Salary"].fillna(Salary_median, inplace=True)

df.head(4)
```

```python
# Renomeando Coluna Salary para Annual_Salary
df.rename(columns={"Salary": "Annual_Salary"},inplace=True)
```

```python
# Criando coluna Month_Salary
linhas = df.shape[0]
month_salary = []

for n in range(linhas):
  month_salary.append(int(df["Annual_Salary"][n])/12)

df.head(3)
```

```python
# Adicioando coluna no data frame
df["Month_Salary"] = month_salary
df.head(3)
```

```python
# Criando Day_Salary
day_salary = []
linhas = df.shape[0]

for n in range(linhas):
  day_salary.append(int(df["Annual_Salary"][n])/360)

df["Day_Salary"] = day_salary
df.head(10)
```

```python
#Alterando coluna Bonus para float

def convert_bonus(i):
  try:
    float(i)
  except:
    np.nan

for n in range(linhas):
  float(df["Bonus %"][n]/100)

df["Bonus %"] = df["Bonus %"].map(convert_bonus)

print(df["Bonus %"].head(5),"\n\n")
df.dtypes
```
